```python
import warnings
from Bio import BiopythonDeprecationWarning

# Suppress warnings about deprecated modules
warnings.filterwarnings("ignore", category=BiopythonDeprecationWarning)
from Bio import SeqIO
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd  # For tabular output

# Your email (for possible API usage later, or for use in headers)
email = "althiindira85@gmail.com"

# Define the file paths to the drug protein FASTA files (excluding VP28)
drug_fasta_paths = {
    "Garlic": r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\garlic.fasta.fasta",
    "HSP70": r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\HSP70.fasta.fasta",
    "Lactobacillus": r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\Lactobacillus.fasta",
    "Lysozymes_c": r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\lysozymes_c.fasta.fasta",
    "Lysozymes_p": r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\lysozymes_P.fasta.fasta",
    "Turmeric": r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\turmeric.fasta.fasta",
     "neem": r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\Neem.fasta.fasta",
     "Horsehoe crab": r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\Horseshoe Crab.fasta.fasta",
     "sea cucumber": r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\Sea Cucumber.fasta.fasta"
}

# Path to the WSSV protein FASTA file
wssv_path = r"C:\Users\Althi\OneDrive\Desktop\WSSV.PRJ\VP28.fasta.fasta"

# Function to retrieve a protein sequence from a FASTA file
def get_protein_sequence(fasta_path):
    for record in SeqIO.parse(fasta_path, "fasta"):
        return str(record.seq), record.id

# Function to perform pairwise alignment and calculate confidence
def calculate_confidence(seq1, seq2):
    alignments = pairwise2.align.globalxx(seq1, seq2)
    score = alignments[0][2] if alignments else 0  # Use the score of the best alignment
    confidence_value = score / max(len(seq1), len(seq2)) if max(len(seq1), len(seq2)) > 0 else 0
    return confidence_value, len(alignments) > 0  # Return confidence value and whether interaction exists

# Create a graph to represent potential interactions
G = nx.Graph()

# List to hold interaction results
interaction_results = []

# Get the WSSV protein sequence
wssv_sequence, wssv_id = get_protein_sequence(wssv_path)

# Calculate interactions for each drug protein
for name, path in drug_fasta_paths.items():
    drug_sequence, drug_id = get_protein_sequence(path)
    confidence_value, interaction_exists = calculate_confidence(wssv_sequence, drug_sequence)

    print(f"Confidence between WSSV ({wssv_id}) and {name}: {confidence_value:.2f}")

    # Append interaction result for tabular output
    interaction_results.append({
        "Drug Protein": name,
        "WSSV Protein ID": wssv_id,
        "Confidence Value": confidence_value,
        "Interaction Exists": interaction_exists
    })

    if interaction_exists:
        # Add nodes and edges to the graph
        G.add_node(f"WSSV ({wssv_id})", color="red")
        G.add_node(f"{name} ({drug_id})", color="blue")
        G.add_edge(f"WSSV ({wssv_id})", f"{name} ({drug_id})", weight=confidence_value)

# Convert results to a DataFrame for tabular display
results_df = pd.DataFrame(interaction_results)

# Display results
print(results_df)

# Plot the interaction network if interactions exist
if G.number_of_edges() > 0:
    plt.figure(figsize=(8, 10))
    pos = nx.spring_layout(G)  # Layout for better visualization
    node_colors = [G.nodes[n]["color"] for n in G.nodes]
    nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=3000, font_size=10)
    plt.title("Protein-Protein Interaction Network between Drug Proteins and WSSV")
    plt.show()
else:
    print("No potential interactions found based on sequence alignment.")
```
